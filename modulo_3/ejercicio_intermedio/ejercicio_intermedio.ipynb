```python
columnas = {col:col.lower() for col in df.columns}
df.rename(columns=columnas, inplace=True)
df.columns
```

```python
# Otra solución 
df.columns = df.columns.str.lower()
df.columns
```

```python
cols = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']
df[cols] = df[cols].apply(lambda x: x.str.replace(',', '.')).astype(float, errors='ignore') # Al astype le puedo pasar el argumento errors= 'ignore'
df[cols]
```

```python
df[cols].dtypes
```

```python
# Este ejercicio se hace de mejor manera con una función con try except para manejar los errores
def cambiar_comas(valColumna):
    try:
        return float(valColumna.replace(",", "."))  
    except:
        return np.nan
    
for columna in cols:
    df[columna] = df[columna].apply(cambiar_comas)
```

```python
# Otra manera de hacerlo mas sofisticada
cols = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']
for col in cols:
    df[col] = df[col].apply(lambda dato: float(dato.replace(",", ".")) if isinstance(dato, str) else np.nan)
```

```python
df[['track', 'title', 'album']] = df[['track', 'title', 'album']].apply(lambda x: x.str.lower())
df[['track', 'title', 'album']]
```

```python
df['artist'] = df['artist'].apply(lambda x: x.replace('$', 's'))
```

```python
# Otra solución
artistas_con_dolar = df['artist'].str.contains('$', regex=False, case = False)
artistas=df[artistas_con_dolar]
artistas["artist"].unique()
```
